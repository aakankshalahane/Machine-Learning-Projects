```python
#E-Commerce Order Analysis
import pandas as pd
orders = pd.DataFrame({
    'Customer': ['Asha', 'Kamal', 'Priya', 'Ravi', 'Asha'],
    'Product': ['Laptop', 'Phone', 'Laptop', 'Tablet', 'Phone'],
    'Amount': [50000, 20000, 55000, 30000, 20000],
    'Quantity': [1, 2, 1, 1, 3]
})
print("Top Customer:", orders.groupby('Customer')['Amount'].sum().idxmax())
print("Most Purchased Product:", orders.groupby('Product')['Quantity'].sum().idxmax())
```
