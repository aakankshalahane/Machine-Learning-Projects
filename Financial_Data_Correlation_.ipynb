```python
#Financial Data Correlation
import pandas as pd
finance = pd.DataFrame({
    'Date': pd.date_range(start='2025-01-01', periods=5),
    'Stock_A': [100, 102, 104, 107, 110],
    'Stock_B': [98, 99, 101, 103, 108]
}).set_index('Date')
print("Correlation Matrix:\n", finance.corr())
```
