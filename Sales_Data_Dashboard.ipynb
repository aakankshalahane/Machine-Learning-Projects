```python
#Sales Data Dashboard
import pandas as pd
sales = pd.DataFrame({
    'Month': ['Jan', 'Feb', 'Mar', 'Jan', 'Feb', 'Mar'],
    'Product': ['A', 'A', 'A', 'B', 'B', 'B'],
    'Revenue': [1000, 1500, 1200, 900, 1600, 1700]
})
print("Total Revenue:", sales['Revenue'].sum())
print("Top Product:", sales.groupby('Product')['Revenue'].sum().idxmax())
print("Monthly Revenue:\n", sales.groupby('Month')['Revenue'].sum())
```
