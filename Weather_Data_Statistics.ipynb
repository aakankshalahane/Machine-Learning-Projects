```python
#Weather Data Statistics
import pandas as pd
import numpy as np # Import numpy

weather = pd.DataFrame({
    'City': ['Mumbai', 'Delhi', 'Pune', 'Chennai'],
    'Temperature': [33, np.nan, 29, 34],
    'Humidity': [70, 50, 60, 80]
})
weather['Temperature'] = weather['Temperature'].fillna(weather['Temperature'].mean())
print(weather)
print("Average Temp per City:\n", weather.groupby('City')['Temperature'].mean())
```
