```python
#Movie Ratings Analyzer
print("\n Movie Ratings Analyzer")
import pandas as pd
movies = pd.DataFrame({
    'Title': ['Movie A', 'Movie B', 'Movie C', 'Movie D'],
    'Genre': ['Action', 'Drama', 'Action', 'Comedy'],
    'Rating': [8.2, 7.5, 9.0, 6.8]
})
print("Average Rating by Genre:\n", movies.groupby('Genre')['Rating'].mean())
print("Top Rated Movie:\n", movies.loc[movies['Rating'].idxmax()])
```
