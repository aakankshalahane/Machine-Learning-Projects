```python
#Employee Salary Insights
import pandas as pd
employees = pd.DataFrame({
    'Name': ['Asha', 'Kamal', 'Priya', 'Ravi', 'Neha'],
    'Department': ['IT', 'HR', 'IT', 'Sales', 'HR'],
    'Salary': [60000, 50000, 75000, 45000, 52000]
})
print("Average Salary by Dept:\n", employees.groupby('Department')['Salary'].mean())
print("Highest Paid Employee:\n", employees.loc[employees['Salary'].idxmax()])
```
