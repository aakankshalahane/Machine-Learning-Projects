```python
#IPL Cricket Stats Analyzer
import pandas as pd
ipl = pd.DataFrame({
    'Player': ['Virat', 'Rohit', 'Dhoni', 'Kohli', 'Hardik'],
    'Team': ['RCB', 'MI', 'CSK', 'RCB', 'MI'],
    'Runs': [78, 56, 44, 100, 65]
})
print("Top Run Scorer:", ipl.groupby('Player')['Runs'].sum().idxmax())
print("Average Runs per Team:\n", ipl.groupby('Team')['Runs'].mean())

```
