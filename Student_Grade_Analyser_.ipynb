```python
import pandas as pd
students = pd.DataFrame({
    'Name': ['Asha', 'Kamal', 'Priya', 'Ravi'],
    'Math': [78, 85, 92, 66],
    'Science': [80, 75, 90, 70],
    'English': [88, 82, 85, 60]
})
students["Average"] = students[['Math', 'Science', 'English']].mean(axis=1)
print(students)
print("Top Student:", students.loc[students['Average'].idxmax(), 'Name'])
```
